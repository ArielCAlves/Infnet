```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score
```

```python
lista_cols = ['Idade','Pontuação teste','Departmento','Distancia casa-trabalho','Educacao','Subordinado','Satisfação com o ambiente no emprego atual','Genero','Cargo','Estado civil','Renda','Quantidade de empresas que trabalho','Anos de experiencia','Anos na última empresa','Contratar']
```

```python
df = pd.read_csv('dataset.csv',usecols=lista_cols)
```

```python
df.head(3)
```

```python
bins = [18, 31, 45, 60]

df['faixa-etaria'] = pd.cut(df['Idade'], bins=bins, labels=['18-30', '31-44', '45-60'])
df.drop(columns=['Idade'],axis=1, inplace=True)
```

```python
bins = [102, 600, 1100, 1500]

df['faixa-pontuacoes'] = pd.cut(df['Pontuação teste'], bins=bins, labels=['Ruim', 'Bom', 'Excelente'])
df.drop(columns=['Pontuação teste'],axis=1, inplace=True)
```

```python
bins = [0, 3, 10, 30]

df['distancia'] = pd.cut(df['Distancia casa-trabalho'], bins=bins, labels=['Perto', 'Médio', 'Longe'])
df.drop(columns=['Distancia casa-trabalho'],axis=1, inplace=True)
```

```python
bins = [0, 20, 100, 2055]

df['equipe'] = pd.cut(df['Subordinado'], bins=bins, labels=['Pequena', 'Média', 'Grande'])
df.drop(columns=['Subordinado'],axis=1, inplace=True)
```

```python
bins = [0, 5000, 10000, 20000]

df['renda'] = pd.cut(df['Renda'], bins=bins, labels=['Baixa', 'Média', 'Alta'])
df.drop(columns=['Renda'],axis=1, inplace=True)
```

```python
df.head(3)
```

```python
df['Contratar'] = df['Contratar'].replace('Não',0).replace('Sim',1)
```

```python
# df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
fig = plt.plot()
sns.countplot(x='Contratar', data=df)
```

```python
# fazer dummie
df2 = pd.get_dummies(df)
```

```python
df2.head()
```

```python
X = df2.drop(['Contratar'], axis=1)
y = df2['Contratar']

#padronizando as colunas numéricas
X = StandardScaler().fit_transform(X)

#label encoder na variável alvo
y = LabelEncoder().fit_transform(y)

#separando treino x teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=133, stratify=y)
```

```python
model = RandomForestClassifier(n_estimators = 100, criterion = 'entropy')
model.fit(X_train, y_train)
```

```python
#previsões
y_pred = model.predict(X_test)

#acurácia do modelo
print('Acurácia: {}%'.format(100*round(accuracy_score(y_test, y_pred),2)))
```

```python
import pickle
with open('./static/model/rfc_model.p', 'wb') as file:
    pickle.dump(model, file)
```
